## Determinación del valor crítico y región de rechazo

### Subtask:
Calcula el valor crítico y define la región de rechazo en función del nivel de significancia y la distribución del estadístico de prueba.


**Reasoning**:
Import the necessary function from scipy.stats for the normal distribution, as this is a common distribution used in hypothesis testing and was mentioned as a possibility in the previous step.



```python
# Importar la función necesaria de scipy.stats para la distribución normal estándar
from scipy.stats import norm
# La distribución normal estándar (Z) es común para pruebas de medias con muestras grandes o desviación estándar poblacional conocida.

# Determinar el tipo de prueba (unilateral o bilateral)
# Esto depende de la hipótesis alternativa (H₁) definida en el paso 1.
# Si H₁ es de la forma μ > μ₀ o μ < μ₀, es una prueba unilateral.
# Si H₁ es de la forma μ ≠ μ₀, es una prueba bilateral.
# Dado que no se especificó una hipótesis alternativa concreta en el paso 1, asumiremos una prueba bilateral para este ejemplo.
tipo_prueba = "bilateral" # O "unilateral_derecha", "unilateral_izquierda"

# Calcular el valor o los valores críticos
# Utilizamos la función ppf (percent point function), que es la inversa de la CDF.

if tipo_prueba == "bilateral":
    # Para una prueba bilateral, necesitamos dos valores críticos que dividen el 5% (alpha) en dos colas del 2.5% (alpha/2) cada una.
    alpha_cola = alpha / 2
    valor_critico_inferior = norm.ppf(alpha_cola)
    valor_critico_superior = norm.ppf(1 - alpha_cola)
    # Documentar los valores críticos
    print(f"Para una prueba bilateral con α = {alpha}:")
    print(f"Valor Crítico Inferior (Z): {valor_critico_inferior:.4f}")
    print(f"Valor Crítico Superior (Z): {valor_critico_superior:.4f}")
    # Definir la región de rechazo
    # La región de rechazo son los valores del estadístico de prueba que caen por debajo del valor crítico inferior o por encima del valor crítico superior.
    print(f"Región de Rechazo: Z < {valor_critico_inferior:.4f} o Z > {valor_critico_superior:.4f}")

elif tipo_prueba == "unilateral_derecha":
    # Para una prueba unilateral derecha, necesitamos un valor crítico que deja el 5% (alpha) en la cola derecha.
    valor_critico = norm.ppf(1 - alpha)
    # Documentar el valor crítico
    print(f"Para una prueba unilateral derecha con α = {alpha}:")
    print(f"Valor Crítico (Z): {valor_critico:.4f}")
    # Definir la región de rechazo
    # La región de rechazo son los valores del estadístico de prueba que caen por encima del valor crítico.
    print(f"Región de Rechazo: Z > {valor_critico:.4f}")

elif tipo_prueba == "unilateral_izquierda":
    # Para una prueba unilateral izquierda, necesitamos un valor crítico que deja el 5% (alpha) en la cola izquierda.
    valor_critico = norm.ppf(alpha)
    # Documentar el valor crítico
    print(f"Para una prueba unilateral izquierda con α = {alpha}:")
    print(f"Valor Crítico (Z): {valor_critico:.4f}")
    # Definir la región de rechazo
    # La región de rechazo son los valores del estadístico de prueba que caen por debajo del valor crítico.
    print(f"Región de Rechazo: Z < {valor_critico:.4f}")

else:
    print("Tipo de prueba no reconocido. Por favor, especifique 'bilateral', 'unilateral_derecha' o 'unilateral_izquierda'.")

# Explicación de la Región de Rechazo: Si el estadístico de prueba calculado cae dentro de esta región, rechazamos la hipótesis nula.
```

## Cálculo del estadístico de prueba y valor p

### Subtask:
Calcula el estadístico de prueba utilizando los datos de la muestra y determina el valor p asociado a este estadístico.


**Reasoning**:
Since no specific data was provided in the previous steps, I will create a sample dataset and then calculate the Z-statistic and the associated p-value for a one-sample Z-test, assuming a known population standard deviation and a hypothesized population mean. I will then print these values with explanations in Spanish.



```python
import numpy as np
from scipy.stats import norm

# 1. Definir un conjunto de datos de ejemplo.
# Asumimos una muestra de tamaño 50 con una media muestral observada.
# Consideramos que la desviación estándar de la población es conocida (sigma).
# Consideramos una media poblacional hipotetizada (mu_0).

np.random.seed(42) # para reproducibilidad
tamano_muestra = 50
media_muestra = 105 # Media observada en nuestra muestra
desviacion_estandar_poblacion = 10 # Desviación estándar conocida de la población
media_poblacional_hipotetizada = 100 # Valor de la media bajo la hipótesis nula

# Generar datos de ejemplo (solo para ilustración, el cálculo real usa la media y desviación estándar)
# datos_ejemplo = np.random.normal(media_muestra, desviacion_estandar_poblacion / np.sqrt(tamano_muestra), tamano_muestra)
# No necesitamos generar los datos reales para calcular el estadístico Z si tenemos la media muestral, tamaño de muestra y desviación estándar poblacional.

# 2. Calcular el estadístico de prueba (Estadístico Z).
# Fórmula para el estadístico Z de una muestra: Z = (x̄ - μ₀) / (σ / √n)
# Donde:
# x̄ es la media de la muestra
# μ₀ es la media poblacional hipotetizada (bajo H₀)
# σ es la desviación estándar de la población
# n es el tamaño de la muestra

estadistico_z = (media_muestra - media_poblacional_hipotetizada) / (desviacion_estandar_poblacion / np.sqrt(tamano_muestra))

# 3. Calcular el valor p asociado al estadístico de prueba.
# Para una prueba bilateral, el valor p es la probabilidad de observar un estadístico de prueba
# tan extremo o más extremo que el calculado en cualquier dirección.
# Es decir, P(Z > |estadistico_z|) + P(Z < -|estadistico_z|).
# Debido a la simetría de la distribución normal estándar, esto es 2 * P(Z > |estadistico_z|)
# o 2 * P(Z < -|estadistico_z|).
# Usamos la función cdf (cumulative distribution function) de scipy.stats.norm
# cdf(z) da P(Z <= z).
# P(Z > z) = 1 - cdf(z)
# P(Z < z) = cdf(z)

# Calculamos la probabilidad en la cola derecha para el valor absoluto del estadístico Z.
valor_p_cola_derecha = 1 - norm.cdf(abs(estadistico_z))

# Para una prueba bilateral, multiplicamos por 2.
valor_p = 2 * valor_p_cola_derecha

# 4. Imprimir el valor del estadístico de prueba y el valor p.
print(f"Estadístico de Prueba Calculado (Z): {estadistico_z:.4f}")
print(f"Explicación del Estadístico Z: Mide cuántas desviaciones estándar la media muestral ({media_muestra}) está de la media poblacional hipotetizada ({media_poblacional_hipotetizada}), considerando la variabilidad muestral.")
print(f"\nValor p Asociado: {valor_p:.4f}")
print(f"Explicación del Valor p: Es la probabilidad de obtener un estadístico de prueba tan extremo o más extremo que el observado ({abs(estadistico_z):.4f}), asumiendo que la hipótesis nula es verdadera. Un valor p pequeño (< α) sugiere evidencia contra H₀.")
```

```python
import numpy as np
import scipy.stats as st

# Datos proporcionados
desviacion_estandar_poblacional = 12  # Desviación estándar poblacional conocida (σ)
margen_error_deseado = 3  # Margen de error deseado
nivel_confianza = 0.90       # Nivel de confianza deseado (90%)

# Nivel de significancia (alpha)
alpha = 1 - nivel_confianza

# 1. Calcular el valor crítico Z para el nivel de confianza del 90%.
# Para un intervalo de confianza bilateral, necesitamos el valor Z que deja (alpha/2) en cada cola.
# Usamos la función ppf (percent point function), que es la inversa de la CDF.
# st.norm.ppf(1 - alpha/2) nos da el valor Z positivo.
z_critico = st.norm.ppf(1 - alpha/2)

# 2. Calcular el tamaño de muestra necesario.
# Fórmula para el tamaño de muestra (medias, σ conocida):
# n = (Z_critico * σ / ME)²
# Donde:
# Z_critico es el valor crítico Z
# σ es la desviación estándar poblacional
# ME es el margen de error deseado

tamano_muestra_necesario = ((z_critico * desviacion_estandar_poblacional) / margen_error_deseado)**2

# Redondear hacia arriba al entero más cercano, ya que el tamaño de muestra debe ser un número entero.
tamano_muestra_necesario = np.ceil(tamano_muestra_necesario)

# 3. Presentar los resultados en español con explicaciones.
print("--- Cálculo del Tamaño de Muestra (para la media, σ conocida) ---")
print("\nDatos Utilizados:")
print(f"  Desviación Estándar Poblacional (σ): {desviacion_estandar_poblacional}")
print(f"  Margen de Error Deseado (ME): {margen_error_deseado}")
print(f"  Nivel de Confianza: {nivel_confianza:.0%}")
print(f"  Valor Crítico Z: {z_critico:.4f}")

print("\nFórmula Utilizada:")
print("n = (Z_critico * σ / ME)²")

print("\nResultado:")
print(f"  Tamaño de muestra necesario: {int(tamano_muestra_necesario)}")
print(f"Explicación: Se necesitarían aproximadamente {int(tamano_muestra_necesario)} estudiantes en la muestra para estimar el puntaje promedio del examen con un margen de error de {margen_error_deseado} puntos y un {nivel_confianza:.0%} de confianza.")
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Datos proporcionados
media_muestra = 495  # Media muestral en gramos
tamano_muestra = 40  # Tamaño de la muestra
desviacion_estandar_poblacional = 15  # Desviación estándar poblacional conocida en gramos (σ)
media_poblacional_hipotetizada = 500  # Media poblacional hipotetizada (bajo H₀) en gramos
alpha = 0.05  # Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa.
# H₀: La máquina está calibrada correctamente (la media poblacional es igual a 500 g). μ = 500
# H₁: La máquina está subllenando (la media poblacional es menor a 500 g). μ < 500
# Esta es una prueba unilateral izquierda.

print("--- Paso 1: Hipótesis ---")
print(f"Hipótesis Nula (H₀): μ = {media_poblacional_hipotetizada} g (La máquina está calibrada correctamente)")
print(f"Hipótesis Alternativa (H₁): μ < {media_poblacional_hipotetizada} g (La máquina está subllenando)")
print(f"Nivel de Significancia (α): {alpha}")

# 2. Calcular el estadístico Z.
# Fórmula para el estadístico Z de una muestra: Z = (x̄ - μ₀) / (σ / √n)
# Donde:
# x̄ es la media de la muestra (`media_muestra`)
# μ₀ es la media poblacional hipotetizada (`media_poblacional_hipotetizada`)
# σ es la desviación estándar de la población (`desviacion_estandar_poblacional`)
# n es el tamaño de la muestra (`tamano_muestra`)

estadistico_z = (media_muestra - media_poblacional_hipotetizada) / (desviacion_estandar_poblacional / np.sqrt(tamano_muestra))

print("\n--- Paso 2: Cálculo del Estadístico Z ---")
print(f"Media Muestral (x̄): {media_muestra} g")
print(f"Tamaño de la Muestra (n): {tamano_muestra}")
print(f"Desviación Estándar Poblacional (σ): {desviacion_estandar_poblacional} g")
print(f"Media Poblacional Hipotetizada (μ₀): {media_poblacional_hipotetizada} g")
print(f"Estadístico Z Calculado: {estadistico_z:.4f}")
print("Explicación: Este valor indica cuántas desviaciones estándar nuestra media muestral está por debajo de la media hipotetizada, considerando la variabilidad muestral.")

# 3. Determinar la región crítica y la región de rechazo.
# Para una prueba unilateral izquierda con α = 0.05, la región de rechazo está en la cola izquierda.
# Necesitamos encontrar el valor crítico Z que deja α en la cola izquierda.
valor_critico_z = st.norm.ppf(alpha)

print("\n--- Paso 3: Región Crítica y Región de Rechazo ---")
print(f"Tipo de Prueba: Unilateral Izquierda")
print(f"Valor Crítico Z para α = {alpha}: {valor_critico_z:.4f}")
print(f"Región de Rechazo: Z < {valor_critico_z:.4f}")
print("Explicación: Si nuestro estadístico Z calculado es menor que el valor crítico, rechazaremos la hipótesis nula.")

# 4. Concluir si se rechaza H0.
# Comparamos el estadístico Z calculado con el valor crítico Z.
decision = ""
if estadistico_z < valor_critico_z:
    decision = "Rechazar la hipótesis nula (H₀)."
else:
    decision = "No rechazar la hipótesis nula (H₀)."

# También podemos calcular el valor p para complementar la decisión.
# Para una prueba unilateral izquierda, el valor p es la probabilidad de observar un estadístico Z
# tan pequeño o más pequeño que el calculado: P(Z <= estadistico_z).
valor_p = st.norm.cdf(estadistico_z)

print("\n--- Paso 4: Decisión y Conclusión ---")
print(f"Estadístico Z Calculado: {estadistico_z:.4f}")
print(f"Valor Crítico Z: {valor_critico_z:.4f}")
print(f"Valor p Asociado: {valor_p:.4f}")
print(f"Decisión Estadística: {decision}")

# Conclusión práctica
print("\nConclusión Práctica:")
if "Rechazar" in decision:
    print(f"Dado que el estadístico Z calculado ({estadistico_z:.4f}) es menor que el valor crítico Z ({valor_critico_z:.4f}) y el valor p ({valor_p:.4f}) es menor que el nivel de significancia ({alpha}), tenemos evidencia estadísticamente significativa para rechazar la hipótesis nula.")
    print(f"Concluimos que hay suficiente evidencia para afirmar que la media poblacional del peso de los paquetes de cereal es significativamente menor a {media_poblacional_hipotetizada} g. Esto sugiere que la máquina de llenado podría estar descalibrada y subllenando los paquetes.")
else:
    print(f"Dado que el estadístico Z calculado ({estadistico_z:.4f}) no es menor que el valor crítico Z ({valor_critico_z:.4f}) y el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({alpha}), no tenemos suficiente evidencia estadística para rechazar la hipótesis nula.")
    print(f"Concluimos que no hay suficiente evidencia para afirmar que la media poblacional del peso de los paquetes de cereal es significativamente menor a {media_poblacional_hipotetizada} g. La diferencia observada en la muestra ({media_muestra} g) podría deberse al azar.")

# 5. Visualización.
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
pdf = st.norm.pdf(x)
plt.plot(x, pdf, label='Distribución Normal Estándar', color='blue')

# Región de rechazo (cola izquierda)
x_rechazo = x[x < valor_critico_z]
pdf_rechazo = pdf[x < valor_critico_z]
plt.fill_between(x_rechazo, 0, pdf_rechazo, color='red', alpha=0.3, label='Región de Rechazo (α = 0.05)')

# Valor crítico Z
plt.axvline(valor_critico_z, color='red', linestyle='--', label=f'Valor Crítico Z ({valor_critico_z:.4f})')

# Estadístico Z calculado
plt.axvline(estadistico_z, color='green', linestyle='-', label=f'Estadístico Z Calculado ({estadistico_z:.4f})')

plt.title('Distribución Normal Estándar con Región de Rechazo y Estadístico de Prueba (Prueba Unilateral Izquierda)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Datos proporcionados
media_muestra = 78  # Media muestral de la clase
tamano_muestra = 36  # Tamaño de la muestra de estudiantes de la clase
desviacion_estandar_poblacional = 12  # Desviación estándar poblacional conocida (σ)
media_poblacional_hipotetizada = 75  # Promedio nacional hipotetizado (bajo H₀)
alpha = 0.05  # Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa.
# H₀: El promedio de la clase es igual al promedio nacional. μ = 75
# H₁: El promedio de la clase es diferente al promedio nacional. μ ≠ 75
# Esta es una prueba bilateral.

print("--- Paso 1: Hipótesis ---")
print(f"Hipótesis Nula (H₀): μ = {media_poblacional_hipotetizada} (El promedio de la clase es igual al promedio nacional)")
print(f"Hipótesis Alternativa (H₁): μ ≠ {media_poblacional_hipotetizada} (El promedio de la clase es diferente al promedio nacional)")
print(f"Nivel de Significancia (α): {alpha}")

# 2. Calcular el estadístico Z.
# Fórmula para el estadístico Z de una muestra: Z = (x̄ - μ₀) / (σ / √n)
# Donde:
# x̄ es la media de la muestra (`media_muestra`)
# μ₀ es la media poblacional hipotetizada (`media_poblacional_hipotetizada`)
# σ es la desviación estándar de la población (`desviacion_estandar_poblacional`)
# n es el tamaño de la muestra (`tamano_muestra`)

estadistico_z = (media_muestra - media_poblacional_hipotetizada) / (desviacion_estandar_poblacional / np.sqrt(tamano_muestra))

print("\n--- Paso 2: Cálculo del Estadístico Z ---")
print(f"Media Muestral (x̄): {media_muestra}")
print(f"Tamaño de la Muestra (n): {tamano_muestra}")
print(f"Desviación Estándar Poblacional (σ): {desviacion_estandar_poblacional}")
print(f"Media Poblacional Hipotetizada (μ₀): {media_poblacional_hipotetizada}")
print(f"Estadístico Z Calculado: {estadistico_z:.4f}")
print("Explicación: Este valor indica cuántas desviaciones estándar nuestra media muestral está de la media hipotetizada, considerando la variabilidad muestral.")

# 3. Determinar la región crítica y la región de rechazo.
# Para una prueba bilateral con α = 0.05, la región de rechazo se divide en dos colas, cada una con α/2.
alpha_cola = alpha / 2
valor_critico_inferior_z = st.norm.ppf(alpha_cola)
valor_critico_superior_z = st.norm.ppf(1 - alpha_cola)


print("\n--- Paso 3: Región Crítica y Región de Rechazo ---")
print(f"Tipo de Prueba: Bilateral")
print(f"Nivel de Significancia (α): {alpha}")
print(f"Valor Crítico Inferior Z: {valor_critico_inferior_z:.4f}")
print(f"Valor Crítico Superior Z: {valor_critico_superior_z:.4f}")
print(f"Región de Rechazo: Z < {valor_critico_inferior_z:.4f} o Z > {valor_critico_superior_z:.4f}")
print("Explicación: Si nuestro estadístico Z calculado cae en una de estas regiones, rechazaremos la hipótesis nula.")

# 4. Concluir si se rechaza H0.
# Comparamos el estadístico Z calculado con los valores críticos Z.
decision = ""
if estadistico_z < valor_critico_inferior_z or estadistico_z > valor_critico_superior_z:
    decision = "Rechazar la hipótesis nula (H₀)."
else:
    decision = "No rechazar la hipótesis nula (H₀)."

# También podemos calcular el valor p para complementar la decisión.
# Para una prueba bilateral, el valor p es la probabilidad de observar un estadístico Z
# tan extremo o más extremo que el calculado en cualquier dirección.
# Es decir, P(Z > |estadistico_z|) + P(Z < -|estadistico_z|) = 2 * P(Z > |estadistico_z|).
valor_p = 2 * (1 - st.norm.cdf(abs(estadistico_z)))


print("\n--- Paso 4: Decisión y Conclusión ---")
print(f"Estadístico Z Calculado: {estadistico_z:.4f}")
print(f"Valores Críticos Z: [{valor_critico_inferior_z:.4f}, {valor_critico_superior_z:.4f}]")
print(f"Valor p Asociado: {valor_p:.4f}")
print(f"Decisión Estadística: {decision}")

# Conclusión práctica
print("\nConclusión Práctica:")
if "Rechazar" in decision:
    print(f"Dado que el valor absoluto del estadístico Z calculado ({abs(estadistico_z):.4f}) es mayor que el valor crítico superior Z ({valor_critico_superior_z:.4f}), o equivalentemente, el valor p ({valor_p:.4f}) es menor que el nivel de significancia ({alpha}), tenemos evidencia estadísticamente significativa para rechazar la hipótesis nula.")
    print(f"Concluimos que hay suficiente evidencia para afirmar que la media poblacional de las calificaciones de la clase del profesor es significativamente diferente del promedio nacional de {media_poblacional_hipotetizada} puntos.")
else:
    print(f"Dado que el valor absoluto del estadístico Z calculado ({abs(estadistico_z):.4f}) no es mayor que el valor crítico superior Z ({valor_critico_superior_z:.4f}), o equivalentemente, el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({alpha}), no tenemos suficiente evidencia estadística para rechazar la hipótesis nula.")
    print(f"Concluimos que no hay suficiente evidencia para afirmar que la media poblacional de las calificaciones de la clase del profesor es significativamente diferente del promedio nacional de {media_poblacional_hipotetizada} puntos. La diferencia observada en la muestra ({media_muestra}) podría deberse al azar.")

# 5. Visualización.
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
pdf = st.norm.pdf(x)
plt.plot(x, pdf, label='Distribución Normal Estándar', color='blue')

# Región de rechazo (dos colas)
x_rechazo_inferior = x[x < valor_critico_inferior_z]
pdf_rechazo_inferior = pdf[x < valor_critico_inferior_z]
plt.fill_between(x_rechazo_inferior, 0, pdf_rechazo_inferior, color='red', alpha=0.3, label=f'Región de Rechazo (α/2 = {alpha_cola})')

x_rechazo_superior = x[x > valor_critico_superior_z]
pdf_rechazo_superior = pdf[x > valor_critico_superior_z]
plt.fill_between(x_rechazo_superior, 0, pdf_rechazo_superior, color='red', alpha=0.3)


# Valores críticos Z
plt.axvline(valor_critico_inferior_z, color='red', linestyle='--', label=f'Valores Críticos Z ({valor_critico_inferior_z:.4f}, {valor_critico_superior_z:.4f})')
plt.axvline(valor_critico_superior_z, color='red', linestyle='--')


# Estadístico Z calculado
plt.axvline(estadistico_z, color='green', linestyle='-', label=f'Estadístico Z Calculado ({estadistico_z:.4f})')

plt.title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico de Prueba (Prueba Bilateral)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Datos proporcionados
media_muestra = 28  # Media muestral del tiempo de entrega en minutos
tamano_muestra = 15  # Tamaño de la muestra de entregas
desviacion_estandar_muestra = 4  # Desviación estándar de la MUESTRA en minutos (s)
media_poblacional_hipotetizada = 25  # Tiempo promedio hipotetizado (bajo H₀) en minutos
alpha = 0.05  # Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa.
# H₀: El tiempo promedio de entrega es 25 minutos. μ = 25
# H₁: El tiempo promedio de entrega es mayor a 25 minutos. μ > 25
# Esta es una prueba unilateral derecha.

print("--- Paso 1: Hipótesis ---")
print(f"Hipótesis Nula (H₀): μ = {media_poblacional_hipotetizada} minutos (El tiempo promedio es 25 minutos)")
print(f"Hipótesis Alternativa (H₁): μ > {media_poblacional_hipotetizada} minutos (El tiempo promedio es mayor a 25 minutos)")
print(f"Nivel de Significancia (α): {alpha}")

# 2. Calcular el estadístico t.
# Fórmula para el estadístico t de una muestra: t = (x̄ - μ₀) / (s / √n)
# Donde:
# x̄ es la media de la muestra (`media_muestra`)
# μ₀ es la media poblacional hipotetizada (`media_poblacional_hipotetizada`)
# s es la desviación estándar de la MUESTRA (`desviacion_estandar_muestra`)
# n es el tamaño de la muestra (`tamano_muestra`)

estadistico_t = (media_muestra - media_poblacional_hipotetizada) / (desviacion_estandar_muestra / np.sqrt(tamano_muestra))

print("\n--- Paso 2: Cálculo del Estadístico t ---")
print(f"Media Muestral (x̄): {media_muestra} minutos")
print(f"Tamaño de la Muestra (n): {tamano_muestra}")
print(f"Desviación Estándar Muestral (s): {desviacion_estandar_muestra} minutos")
print(f"Media Poblacional Hipotetizada (μ₀): {media_poblacional_hipotetizada} minutos")
print(f"Estadístico t Calculado: {estadistico_t:.4f}")
print("Explicación: Este valor indica cuántas desviaciones estándar muestrales nuestra media muestral está por encima de la media hipotetizada, considerando la variabilidad muestral y el tamaño de la muestra.")


# 3. Determinar los grados de libertad y el valor crítico t.
grados_libertad = tamano_muestra - 1

# Para una prueba unilateral derecha con α = 0.05, necesitamos encontrar el valor crítico t que deja α en la cola derecha.
valor_critico_t = st.t.ppf(1 - alpha, df=grados_libertad)

print("\n--- Paso 3: Grados de Libertad, Valor Crítico t y Región de Rechazo ---")
print(f"Grados de Libertad (df = n-1): {grados_libertad}")
print(f"Tipo de Prueba: Unilateral Derecha")
print(f"Valor Crítico t para α = {alpha} y df = {grados_libertad}: {valor_critico_t:.4f}")
print(f"Región de Rechazo: t > {valor_critico_t:.4f}")
print("Explicación: Si nuestro estadístico t calculado es mayor que el valor crítico, rechazaremos la hipótesis nula.")

# 4. Concluir si se rechaza H0.
# Comparamos el estadístico t calculado con el valor crítico t.
decision = ""
if estadistico_t > valor_critico_t:
    decision = "Rechazar la hipótesis nula (H₀)."
else:
    decision = "No rechazar la hipótesis nula (H₀)."

# También podemos calcular el valor p para complementar la decisión.
# Para una prueba unilateral derecha, el valor p es la probabilidad de observar un estadístico t
# tan grande o más grande que el calculado: P(t >= estadistico_t).
# Usamos la función sf (survival function), que es 1 - cdf.
valor_p = st.t.sf(estadistico_t, df=grados_libertad)


print("\n--- Paso 4: Decisión y Conclusión ---")
print(f"Estadístico t Calculado: {estadistico_t:.4f}")
print(f"Valor Crítico t: {valor_critico_t:.4f}")
print(f"Valor p Asociado: {valor_p:.4f}")
print(f"Decisión Estadística: {decision}")

# Conclusión práctica
print("\nConclusión Práctica:")
if "Rechazar" in decision:
    print(f"Dado que el estadístico t calculado ({estadistico_t:.4f}) es mayor que el valor crítico t ({valor_critico_t:.4f}) y el valor p ({valor_p:.4f}) es menor que el nivel de significancia ({alpha}), tenemos evidencia estadísticamente significativa para rechazar la hipótesis nula.")
    print(f"Concluimos que hay suficiente evidencia para afirmar que el verdadero tiempo promedio de entrega a domicilio del restaurante es significativamente mayor a los {media_poblacional_hipotetizada} minutos afirmados.")
else:
    print(f"Dado que el estadístico t calculado ({estadistico_t:.4f}) no es mayor que el valor crítico t ({valor_critico_t:.4f}) y el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({alpha}), no tenemos suficiente evidencia estadística para rechazar la hipótesis nula.")
    print(f"Concluimos que no hay suficiente evidencia para afirmar que el verdadero tiempo promedio de entrega a domicilio del restaurante es significativamente mayor a los {media_poblacional_hipotetizada} minutos. La diferencia observada en la muestra ({media_muestra} minutos) podría deberse al azar.")


# 5. Visualización.
# Generar valores x para la distribución t
x = np.linspace(st.t.ppf(0.001, df=grados_libertad), st.t.ppf(0.999, df=grados_libertad), 1000)
pdf = st.t.pdf(x, df=grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Distribución t (df = {grados_libertad})', color='blue')

# Región de rechazo (cola derecha)
x_rechazo = x[x > valor_critico_t]
pdf_rechazo = pdf[x > valor_critico_t]
plt.fill_between(x_rechazo, 0, pdf_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Valor crítico t
plt.axvline(valor_critico_t, color='red', linestyle='--', label=f'Valor Crítico t ({valor_critico_t:.4f})')

# Estadístico t calculado
plt.axvline(estadistico_t, color='green', linestyle='-', label=f'Estadístico t Calculado ({estadistico_t:.4f})')

plt.title(f'Distribución t (df = {grados_libertad}) con Región de Rechazo y Estadístico de Prueba (Prueba Unilateral Derecha)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Datos proporcionados
media_muestra = 75  # Media muestral de la clase en matemáticas
tamano_muestra = 12  # Tamaño de la muestra de estudiantes
desviacion_estandar_muestra = 8  # Desviación estándar de la MUESTRA
media_poblacional_hipotetizada = 80  # Promedio hipotetizado (bajo H₀)
alpha = 0.01  # Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa.
# H₀: El promedio de la clase en matemáticas es 80. μ = 80
# H₁: El promedio de la clase en matemáticas es diferente de 80. μ ≠ 80
# Esta es una prueba bilateral.

print("--- Paso 1: Hipótesis ---")
print(f"Hipótesis Nula (H₀): μ = {media_poblacional_hipotetizada} (El promedio de la clase en matemáticas es 80)")
print(f"Hipótesis Alternativa (H₁): μ ≠ {media_poblacional_hipotetizada} (El promedio de la clase en matemáticas es diferente de 80)")
print(f"Nivel de Significancia (α): {alpha}")

# 2. Calcular el estadístico t.
# Fórmula para el estadístico t de una muestra: t = (x̄ - μ₀) / (s / √n)
# Donde:
# x̄ es la media de la muestra (`media_muestra`)
# μ₀ es la media poblacional hipotetizada (`media_poblacional_hipotetizada`)
# s es la desviación estándar de la MUESTRA (`desviacion_estandar_muestra`)
# n es el tamaño de la muestra (`tamano_muestra`)

estadistico_t = (media_muestra - media_poblacional_hipotetizada) / (desviacion_estandar_muestra / np.sqrt(tamano_muestra))

print("\n--- Paso 2: Cálculo del Estadístico t ---")
print(f"Media Muestral (x̄): {media_muestra}")
print(f"Tamaño de la Muestra (n): {tamano_muestra}")
print(f"Desviación Estándar Muestral (s): {desviacion_estandar_muestra}")
print(f"Media Poblacional Hipotetizada (μ₀): {media_poblacional_hipotetizada}")
print(f"Estadístico t Calculado: {estadistico_t:.4f}")
print("Explicación: Este valor indica cuántas desviaciones estándar muestrales nuestra media muestral está de la media hipotetizada, considerando la variabilidad muestral y el tamaño de la muestra.")


# 3. Determinar los grados de libertad y los valores críticos t.
grados_libertad = tamano_muestra - 1

# Para una prueba bilateral con α = 0.01, la región de rechazo se divide en dos colas, cada una con α/2.
alpha_cola = alpha / 2
valor_critico_inferior_t = st.t.ppf(alpha_cola, df=grados_libertad)
valor_critico_superior_t = st.t.ppf(1 - alpha_cola, df=grados_libertad)

print("\n--- Paso 3: Grados de Libertad, Valores Críticos t y Regiones de Rechazo ---")
print(f"Grados de Libertad (df = n-1): {grados_libertad}")
print(f"Tipo de Prueba: Bilateral")
print(f"Nivel de Significancia (α): {alpha}")
print(f"Valor Crítico Inferior t para α/2 = {alpha_cola:.4f} y df = {grados_libertad}: {valor_critico_inferior_t:.4f}")
print(f"Valor Crítico Superior t para α/2 = {alpha_cola:.4f} y df = {grados_libertad}: {valor_critico_superior_t:.4f}")
print(f"Región de Rechazo: t < {valor_critico_inferior_t:.4f} o t > {valor_critico_superior_t:.4f}")
print("Explicación: Si nuestro estadístico t calculado cae en una de estas regiones, rechazaremos la hipótesis nula.")

# 4. Concluir si se rechaza H0.
# Comparamos el estadístico t calculado con los valores críticos t.
decision = ""
if estadistico_t < valor_critico_inferior_t or estadistico_t > valor_critico_superior_t:
    decision = "Rechazar la hipótesis nula (H₀)."
else:
    decision = "No rechazar la hipótesis nula (H₀)."

# También podemos calcular el valor p para complementar la decisión.
# Para una prueba bilateral, el valor p es la probabilidad de observar un estadístico t
# tan extremo o más extremo que el calculado en cualquier dirección.
# Es decir, P(t > |estadistico_t|) + P(t < -|estadistico_t|) = 2 * P(t > |estadistico_t|).
valor_p = 2 * st.t.sf(abs(estadistico_t), df=grados_libertad)


print("\n--- Paso 4: Decisión y Conclusión ---")
print(f"Estadístico t Calculado: {estadistico_t:.4f}")
print(f"Valores Críticos t: [{valor_critico_inferior_t:.4f}, {valor_critico_superior_t:.4f}]")
print(f"Valor p Asociado: {valor_p:.4f}")
print(f"Decisión Estadística: {decision}")

# Conclusión práctica
print("\nConclusión Práctica:")
if "Rechazar" in decision:
    print(f"Dado que el valor absoluto del estadístico t calculado ({abs(estadistico_t):.4f}) es mayor que el valor crítico superior t ({valor_critico_superior_t:.4f}), o equivalentemente, el valor p ({valor_p:.4f}) es menor que el nivel de significancia ({alpha}), tenemos evidencia estadísticamente significativa para rechazar la hipótesis nula.")
    print(f"Concluimos que hay suficiente evidencia para afirmar que la media poblacional de las calificaciones de la clase en matemáticas es significativamente diferente de {media_poblacional_hipotetizada}.")
else:
    print(f"Dado que el valor absoluto del estadístico t calculado ({abs(estadistico_t):.4f}) no es mayor que el valor crítico superior t ({valor_critico_superior_t:.4f}), o equivalentemente, el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({alpha}), no tenemos suficiente evidencia estadística para rechazar la hipótesis nula.")
    print(f"Concluimos que no hay suficiente evidencia para afirmar que la media poblacional de las calificaciones de la clase en matemáticas es significativamente diferente de {media_poblacional_hipotetizada}. La diferencia observada en la muestra ({media_muestra}) podría deberse al azar.")


# 5. Visualización.
# Generar valores x para la distribución t
x = np.linspace(st.t.ppf(0.001, df=grados_libertad), st.t.ppf(0.999, df=grados_libertad), 1000)
pdf = st.t.pdf(x, df=grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Distribución t (df = {grados_libertad})', color='blue')

# Regiones de rechazo (dos colas)
x_rechazo_inferior = x[x < valor_critico_inferior_t]
pdf_rechazo_inferior = pdf[x < valor_critico_inferior_t]
plt.fill_between(x_rechazo_inferior, 0, pdf_rechazo_inferior, color='red', alpha=0.3, label=f'Región de Rechazo (α/2 = {alpha_cola})')

x_rechazo_superior = x[x > valor_critico_superior_t]
pdf_rechazo_superior = pdf[x > valor_critico_superior_t]
plt.fill_between(x_rechazo_superior, 0, pdf_rechazo_superior, color='red', alpha=0.3)

# Valores críticos t
plt.axvline(valor_critico_inferior_t, color='red', linestyle='--', label=f'Valores Críticos t ({valor_critico_inferior_t:.4f}, {valor_critico_superior_t:.4f})')
plt.axvline(valor_critico_superior_t, color='red', linestyle='--')

# Estadístico t calculado
plt.axvline(estadistico_t, color='green', linestyle='-', label=f'Estadístico t Calculado ({estadistico_t:.4f})')

plt.title(f'Distribución t (df = {grados_libertad}) con Regiones de Rechazo y Estadístico de Prueba (Prueba Bilateral)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Datos proporcionados
varianza_poblacional_hipotetizada = 2.5  # Varianza poblacional hipotetizada (σ² bajo H₀)
tamano_muestra = 16  # Tamaño de la muestra
varianza_muestral = 4  # Varianza muestral (s²)
alpha = 0.05  # Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa.
# H₀: La varianza poblacional es igual a 2.5. σ² = 2.5
# H₁: La varianza poblacional es diferente de 2.5. σ² ≠ 2.5
# Esta es una prueba bilateral.

print("--- Paso 1: Hipótesis ---")
print(f"Hipótesis Nula (H₀): σ² = {varianza_poblacional_hipotetizada} minutos² (La varianza poblacional es 2.5)")
print(f"Hipótesis Alternativa (H₁): σ² ≠ {varianza_poblacional_hipotetizada} minutos² (La varianza poblacional es diferente de 2.5)")
print(f"Nivel de Significancia (α): {alpha}")

# 2. Calcule el estadístico χ².
# Fórmula para el estadístico chi-cuadrado para la varianza: χ² = ((n - 1) * s²) / σ²
# Donde:
# n es el tamaño de la muestra (`tamano_muestra`)
# s² es la varianza muestral (`varianza_muestral`)
# σ² es la varianza poblacional hipotetizada (`varianza_poblacional_hipotetizada`)

estadistico_chi2 = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_hipotetizada

print("\n--- Paso 2: Cálculo del Estadístico χ² ---")
print(f"Tamaño de la Muestra (n): {tamano_muestra}")
print(f"Varianza Muestral (s²): {varianza_muestral} minutos²")
print(f"Varianza Poblacional Hipotetizada (σ²): {varianza_poblacional_hipotetizada} minutos²")
print(f"Estadístico χ² Calculado: {estadistico_chi2:.4f}")
print("Explicación: Este valor mide la discrepancia entre la varianza muestral observada y la varianza poblacional hipotetizada.")

# 3. Determine los grados de libertad y los valores críticos χ².
grados_libertad = tamano_muestra - 1

# Para una prueba bilateral con α = 0.05, la región de rechazo se divide en dos colas, cada una con α/2.
alpha_cola = alpha / 2

# Valor crítico inferior: deja α/2 en la cola izquierda.
valor_critico_inferior_chi2 = st.chi2.ppf(alpha_cola, df=grados_libertad)

# Valor crítico superior: deja α/2 en la cola derecha (o 1 - α/2 en la cola izquierda acumulada).
valor_critico_superior_chi2 = st.chi2.ppf(1 - alpha_cola, df=grados_libertad)

print("\n--- Paso 3: Grados de Libertad, Valores Críticos χ² y Regiones de Rechazo ---")
print(f"Grados de Libertad (df = n-1): {grados_libertad}")
print(f"Tipo de Prueba: Bilateral")
print(f"Nivel de Significancia (α): {alpha}")
print(f"Valor Crítico Inferior χ² para α/2 = {alpha_cola:.4f} y df = {grados_libertad}: {valor_critico_inferior_chi2:.4f}")
print(f"Valor Crítico Superior χ² para α/2 = {alpha_cola:.4f} y df = {grados_libertad}: {valor_critico_superior_chi2:.4f}")
print(f"Región de Rechazo: χ² < {valor_critico_inferior_chi2:.4f} o χ² > {valor_critico_superior_chi2:.4f}")
print("Explicación: Si nuestro estadístico χ² calculado cae en una de estas regiones, rechazaremos la hipótesis nula.")

# 4. Concluya si se rechaza H0.
# Comparamos el estadístico χ² calculado con los valores críticos χ².
decision = ""
if estadistico_chi2 < valor_critico_inferior_chi2 or estadistico_chi2 > valor_critico_superior_chi2:
    decision = "Rechazar la hipótesis nula (H₀)."
else:
    decision = "No rechazar la hipótesis nula (H₀)."

# También podemos calcular el valor p para complementar la decisión.
# Para una prueba bilateral, el valor p es la probabilidad de observar un estadístico χ²
# tan extremo o más extremo que el calculado en cualquier dirección.
# Esto implica calcular la probabilidad en la cola más cercana al estadístico de prueba
# y multiplicar por 2.
if estadistico_chi2 < grados_libertad: # Si el estadístico está a la izquierda de la media (grados_libertad)
    valor_p = 2 * st.chi2.cdf(estadistico_chi2, df=grados_libertad)
else: # Si el estadístico está a la derecha de la media
    valor_p = 2 * st.chi2.sf(estadistico_chi2, df=grados_libertad)

# Ajustar el valor p si es mayor que 1 debido a la simetría (puede ocurrir con valores cercanos a la media)
valor_p = min(valor_p, 1)


print("\n--- Paso 4: Decisión y Conclusión ---")
print(f"Estadístico χ² Calculado: {estadistico_chi2:.4f}")
print(f"Valores Críticos χ²: [{valor_critico_inferior_chi2:.4f}, {valor_critico_superior_chi2:.4f}]")
print(f"Valor p Asociado: {valor_p:.4f}")
print(f"Decisión Estadística: {decision}")

# Conclusión práctica
print("\nConclusión Práctica:")
if "Rechazar" in decision:
    print(f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) cae en la región de rechazo (fuera del rango [{valor_critico_inferior_chi2:.4f}, {valor_critico_superior_chi2:.4f}]) y el valor p ({valor_p:.4f}) es menor que el nivel de significancia ({alpha}), tenemos evidencia estadísticamente significativa para rechazar la hipótesis nula.")
    print(f"Concluimos que hay suficiente evidencia para afirmar que la verdadera varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de {varianza_poblacional_hipotetizada} minutos².")
else:
    print(f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) no cae en la región de rechazo (está dentro del rango [{valor_critico_inferior_chi2:.4f}, {valor_critico_superior_chi2:.4f}]) y el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({alpha}), no tenemos suficiente evidencia estadística para rechazar la hipótesis nula.")
    print(f"Concluimos que no hay suficiente evidencia para afirmar que la verdadera varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de {varianza_poblacional_hipotetizada} minutos². La diferencia observada en la muestra podría deberse al azar.")


# 5. Visualización.
# Generar valores x para la distribución chi-cuadrado
# La distribución chi-cuadrado es asimétrica y solo toma valores no negativos.
# El rango típico es de 0 hasta un valor que cubra la mayor parte de la distribución.
# Un buen límite superior puede ser el valor crítico superior + un poco más, o simplemente un valor razonable como 30 o 40 para df=15.
x = np.linspace(0, st.chi2.ppf(0.999, df=grados_libertad) + 5, 500) # Empieza en 0, va hasta un valor adecuado
pdf = st.chi2.pdf(x, df=grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Distribución χ² (df = {grados_libertad})', color='blue')

# Regiones de rechazo (dos colas)
x_rechazo_inferior = x[x < valor_critico_inferior_chi2]
pdf_rechazo_inferior = pdf[x < valor_critico_inferior_chi2]
plt.fill_between(x_rechazo_inferior, 0, pdf_rechazo_inferior, color='red', alpha=0.3, label=f'Región de Rechazo (α/2 = {alpha_cola})')

x_rechazo_superior = x[x > valor_critico_superior_chi2]
pdf_rechazo_superior = pdf[x > valor_critico_superior_chi2]
plt.fill_between(x_rechazo_superior, 0, pdf_rechazo_superior, color='red', alpha=0.3)

# Valores críticos χ²
plt.axvline(valor_critico_inferior_chi2, color='red', linestyle='--', label=f'Valores Críticos χ² ({valor_critico_inferior_chi2:.4f}, {valor_critico_superior_chi2:.4f})')
plt.axvline(valor_critico_superior_chi2, color='red', linestyle='--')

# Estadístico χ² calculado
plt.axvline(estadistico_chi2, color='green', linestyle='-', label=f'Estadístico χ² Calculado ({estadistico_chi2:.4f})')

plt.title(f'Distribución χ² (df = {grados_libertad}) con Regiones de Rechazo y Estadístico de Prueba (Prueba Bilateral de Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Datos proporcionados
varianza_poblacional_hipotetizada = 1.2  # Varianza poblacional hipotetizada (σ² bajo H₀)
tamano_muestra = 20  # Tamaño de la muestra
varianza_muestral = 2.0  # Varianza muestral (s²)
alpha = 0.01  # Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa.
# H₀: La varianza poblacional es igual a 1.2. σ² = 1.2
# H₁: La varianza poblacional es mayor a 1.2. σ² > 1.2
# Esta es una prueba unilateral derecha.

print("--- Paso 1: Hipótesis ---")
print(f"Hipótesis Nula (H₀): σ² = {varianza_poblacional_hipotetizada} ml² (La varianza poblacional es 1.2)")
print(f"Hipótesis Alternativa (H₁): σ² > {varianza_poblacional_hipotetizada} ml² (La varianza poblacional es mayor a 1.2)")
print(f"Nivel de Significancia (α): {alpha}")

# 2. Calcule el estadístico χ².
# Fórmula para el estadístico chi-cuadrado para la varianza: χ² = ((n - 1) * s²) / σ²
# Donde:
# n es el tamaño de la muestra (`tamano_muestra`)
# s² es la varianza muestral (`varianza_muestral`)
# σ² es la varianza poblacional hipotetizada (`varianza_poblacional_hipotetizada`)

estadistico_chi2 = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_hipotetizada

print("\n--- Paso 2: Cálculo del Estadístico χ² ---")
print(f"Tamaño de la Muestra (n): {tamano_muestra}")
print(f"Varianza Muestral (s²): {varianza_muestral} ml²")
print(f"Varianza Poblacional Hipotetizada (σ²): {varianza_poblacional_hipotetizada} ml²")
print(f"Estadístico χ² Calculado: {estadistico_chi2:.4f}")
print("Explicación: Este valor mide la discrepancia entre la varianza muestral observada y la varianza poblacional hipotetizada.")

# 3. Determine los grados de libertad y el valor crítico χ².
grados_libertad = tamano_muestra - 1

# Para una prueba unilateral derecha con α = 0.01, necesitamos encontrar el valor crítico χ² que deja α en la cola derecha.
valor_critico_chi2 = st.chi2.ppf(1 - alpha, df=grados_libertad)

print("\n--- Paso 3: Grados de Libertad, Valor Crítico χ² y Región de Rechazo ---")
print(f"Grados de Libertad (df = n-1): {grados_libertad}")
print(f"Tipo de Prueba: Unilateral Derecha")
print(f"Valor Crítico χ² para α = {alpha} y df = {grados_libertad}: {valor_critico_chi2:.4f}")
print(f"Región de Rechazo: χ² > {valor_critico_chi2:.4f}")
print("Explicación: Si nuestro estadístico χ² calculado es mayor que el valor crítico, rechazaremos la hipótesis nula.")

# 4. Concluya si se rechaza H0.
# Comparamos el estadístico χ² calculado con el valor crítico χ².
decision = ""
if estadistico_chi2 > valor_critico_chi2:
    decision = "Rechazar la hipótesis nula (H₀)."
else:
    decision = "No rechazar la hipótesis nula (H₀)."

# También podemos calcular el valor p para complementar la decisión.
# Para una prueba unilateral derecha, el valor p es la probabilidad de observar un estadístico χ²
# tan grande o más grande que el calculado: P(χ² >= estadistico_chi2).
# Usamos la función sf (survival function), que es 1 - cdf.
valor_p = st.chi2.sf(estadistico_chi2, df=grados_libertad)


print("\n--- Paso 4: Decisión y Conclusión ---")
print(f"Estadístico χ² Calculado: {estadistico_chi2:.4f}")
print(f"Valor Crítico χ²: {valor_critico_chi2:.4f}")
print(f"Valor p Asociado: {valor_p:.4f}")
print(f"Decisión Estadística: {decision}")

# Conclusión práctica
print("\nConclusión Práctica:")
if "Rechazar" in decision:
    print(f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) es mayor que el valor crítico χ² ({valor_critico_chi2:.4f}) y el valor p ({valor_p:.4f}) es menor que el nivel de significancia ({alpha}), tenemos evidencia estadísticamente significativa para rechazar la hipótesis nula.")
    print(f"Concluimos que hay suficiente evidencia para afirmar que la verdadera varianza poblacional del llenado de las botellas es significativamente mayor a {varianza_poblacional_hipotetizada} ml². Esto sugiere que el proceso de llenado tiene más variabilidad de la permitida.")
else:
    print(f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) no es mayor que el valor crítico χ² ({valor_critico_chi2:.4f}) y el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({alpha}), no tenemos suficiente evidencia estadística para rechazar la hipótesis nula.")
    print(f"Concluimos que no hay suficiente evidencia para afirmar que la verdadera varianza poblacional del llenado de las botellas es significativamente mayor a {varianza_poblacional_hipotetizada} ml². La diferencia observada en la muestra ({varianza_muestral} ml²) podría deberse al azar.")


# 5. Visualización.
# Generar valores x para la distribución chi-cuadrado
# La distribución chi-cuadrado es asimétrica y solo toma valores no negativos.
# El rango típico es de 0 hasta un valor que cubra la mayor parte de la distribución.
x = np.linspace(0, st.chi2.ppf(0.999, df=grados_libertad) + 5, 500) # Empieza en 0, va hasta un valor adecuado
pdf = st.chi2.pdf(x, df=grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Distribución χ² (df = {grados_libertad})', color='blue')

# Región de rechazo (cola derecha)
x_rechazo = x[x > valor_critico_chi2]
pdf_rechazo = pdf[x > valor_critico_chi2]
plt.fill_between(x_rechazo, 0, pdf_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Valor crítico χ²
plt.axvline(valor_critico_chi2, color='red', linestyle='--', label=f'Valor Crítico χ² ({valor_critico_chi2:.4f})')

# Estadístico χ² calculado
plt.axvline(estadistico_chi2, color='green', linestyle='-', label=f'Estadístico χ² Calculado ({estadistico_chi2:.4f})')

plt.title(f'Distribución χ² (df = {grados_libertad}) con Región de Rechazo y Estadístico de Prueba (Prueba Unilateral Derecha de Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Datos proporcionados
varianza_poblacional_hipotetizada = 0.5  # Varianza poblacional hipotetizada (σ² bajo H₀)
tamano_muestra = 12  # Tamaño de la muestra
varianza_muestral = 0.3  # Varianza muestral (s²)
alpha = 0.05  # Nivel de significancia

# 1. Plantear las hipótesis nula y alternativa.
# H₀: La varianza poblacional es igual a 0.5. σ² = 0.5
# H₁: La varianza poblacional es menor a 0.5. σ² < 0.5
# Esta es una prueba unilateral izquierda.

print("--- Paso 1: Hipótesis ---")
print(f"Hipótesis Nula (H₀): σ² = {varianza_poblacional_hipotetizada} mm² (La varianza poblacional es 0.5)")
print(f"Hipótesis Alternativa (H₁): σ² < {varianza_poblacional_hipotetizada} mm² (La varianza poblacional es menor a 0.5)")
print(f"Nivel de Significancia (α): {alpha}")

# 2. Calcule el estadístico χ².
# Fórmula para el estadístico chi-cuadrado para la varianza: χ² = ((n - 1) * s²) / σ²
# Donde:
# n es el tamaño de la muestra (`tamano_muestra`)
# s² es la varianza muestral (`varianza_muestral`)
# σ² es la varianza poblacional hipotetizada (`varianza_poblacional_hipotetizada`)

estadistico_chi2 = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_hipotetizada

print("\n--- Paso 2: Cálculo del Estadístico χ² ---")
print(f"Tamaño de la Muestra (n): {tamano_muestra}")
print(f"Varianza Muestral (s²): {varianza_muestral} mm²")
print(f"Varianza Poblacional Hipotetizada (σ²): {varianza_poblacional_hipotetizada} mm²")
print(f"Estadístico χ² Calculado: {estadistico_chi2:.4f}")
print("Explicación: Este valor mide la discrepancia entre la varianza muestral observada y la varianza poblacional hipotetizada.")

# 3. Determine los grados de libertad y el valor crítico χ².
grados_libertad = tamano_muestra - 1

# Para una prueba unilateral izquierda con α = 0.05, necesitamos encontrar el valor crítico χ² que deja α en la cola izquierda.
valor_critico_chi2 = st.chi2.ppf(alpha, df=grados_libertad)

print("\n--- Paso 3: Grados de Libertad, Valor Crítico χ² y Región de Rechazo ---")
print(f"Grados de Libertad (df = n-1): {grados_libertad}")
print(f"Tipo de Prueba: Unilateral Izquierda")
print(f"Valor Crítico χ² para α = {alpha} y df = {grados_libertad}: {valor_critico_chi2:.4f}")
print(f"Región de Rechazo: χ² < {valor_critico_chi2:.4f}")
print("Explicación: Si nuestro estadístico χ² calculado es menor que el valor crítico, rechazaremos la hipótesis nula.")

# 4. Concluya si se rechaza H0.
# Comparamos el estadístico χ² calculado con el valor crítico χ².
decision = ""
if estadistico_chi2 < valor_critico_chi2:
    decision = "Rechazar la hipótesis nula (H₀)."
else:
    decision = "No rechazar la hipótesis nula (H₀)."

# También podemos calcular el valor p para complementar la decisión.
# Para una prueba unilateral izquierda, el valor p es la probabilidad de observar un estadístico χ²
# tan pequeño o más pequeño que el calculado: P(χ² <= estadistico_chi2).
valor_p = st.chi2.cdf(estadistico_chi2, df=grados_libertad)


print("\n--- Paso 4: Decisión y Conclusión ---")
print(f"Estadístico χ² Calculado: {estadistico_chi2:.4f}")
print(f"Valor Crítico χ²: {valor_critico_chi2:.4f}")
print(f"Valor p Asociado: {valor_p:.4f}")
print(f"Decisión Estadística: {decision}")

# Conclusión práctica
print("\nConclusión Práctica:")
if "Rechazar" in decision:
    print(f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) es menor que el valor crítico χ² ({valor_critico_chi2:.4f}) y el valor p ({valor_p:.4f}) es menor que el nivel de significancia ({alpha}), tenemos evidencia estadísticamente significativa para rechazar la hipótesis nula.")
    print(f"Concluimos que hay suficiente evidencia para afirmar que la verdadera varianza poblacional del grosor de las piezas metálicas es significativamente menor a {varianza_poblacional_hipotetizada} mm², lo que sugiere que el proceso de fabricación es menos variable de lo que el control de calidad afirma.")
else:
    print(f"Dado que el estadístico χ² calculado ({estadistico_chi2:.4f}) no es menor que el valor crítico χ² ({valor_critico_chi2:.4f}) y el valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({alpha}), no tenemos suficiente evidencia estadística para rechazar la hipótesis nula.")
    print(f"Concluimos que no hay suficiente evidencia para afirmar que la verdadera varianza poblacional del grosor de las piezas metálicas es significativamente menor a {varianza_poblacional_hipotetizada} mm². La diferencia observada en la muestra ({varianza_muestral} mm²) podría deberse al azar.")

# 5. Visualización.
# Generar valores x para la distribución chi-cuadrado
# La distribución chi-cuadrado es asimétrica y solo toma valores no negativos.
# El rango típico es de 0 hasta un valor que cubra la mayor parte de la distribución.
x = np.linspace(0, st.chi2.ppf(0.999, df=grados_libertad) + 5, 500) # Empieza en 0, va hasta un valor adecuado
pdf = st.chi2.pdf(x, df=grados_libertad)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Distribución χ² (df = {grados_libertad})', color='blue')

# Región de rechazo (cola izquierda)
x_rechazo = x[x < valor_critico_chi2]
pdf_rechazo = pdf[x < valor_critico_chi2]
plt.fill_between(x_rechazo, 0, pdf_rechazo, color='red', alpha=0.3, label=f'Región de Rechazo (α = {alpha})')

# Valor crítico χ²
plt.axvline(valor_critico_chi2, color='red', linestyle='--', label=f'Valor Crítico χ² ({valor_critico_chi2:.4f})')

# Estadístico χ² calculado
plt.axvline(estadistico_chi2, color='green', linestyle='-', label=f'Estadístico χ² Calculado ({estadistico_chi2:.4f})')

plt.title(f'Distribución χ² (df = {grados_libertad}) con Región de Rechazo y Estadístico de Prueba (Prueba Unilateral Izquierda de Varianza)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
